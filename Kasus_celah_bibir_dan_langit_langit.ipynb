```python
data2 = data.dropna()
print(data2.info())
```

```python
from scipy.stats import fisher_exact

Status_penderita = 'Penderita'
variables = [col for col in data2.columns if col!=Status_penderita]
result = []
for var in variables :
  a = ((data2[var]=='Ya')&(data2[Status_penderita]=='Ya')).sum()
  b = ((data2[var]=='Tidak')&(data2[Status_penderita]=='Ya')).sum()
  c = ((data2[var]=='Ya')&(data2[Status_penderita]=='Tidak')).sum()
  d = ((data2[var]=='Tidak')&(data2[Status_penderita]=='Tidak')).sum()
  table = [[a, b],[c, d]]
  oddratio, p= fisher_exact(table, alternative='two-sided')
  result.append({'variables': var, 'a':a, 'b':b, 'c':c, 'd':d, 'odd_ratio':oddratio, 'p_value':p})

fisher_results_df = pd.DataFrame(result)
print(fisher_results_df)
```
